### df_adresses

```python

def adresses(dfindex, lien_avant_objects) :

    liste_adresses=[]

    for i in range(len(df1)) :
        with open(f"{lien_avant_objects}/objects/{df1['file'][i]}", "r") as fichier:
            contenu = fichier.read()

            # Chargement des données JSON
            data = json.loads(contenu)

            try:
                adresse= ", ".join(data['isLocatedAt'][0]['schema:address'][0]['schema:streetAddress'])
                cp = data['isLocatedAt'][0]['schema:address'][0]['schema:postalCode']
                ville = data['isLocatedAt'][0]['schema:address'][0]['schema:addressLocality']
                departement = data['isLocatedAt'][0]['schema:address'][0]['hasAddressCity']['isPartOfDepartment']['rdfs:label']['fr']
                code_dpt = data['isLocatedAt'][0]['schema:address'][0]['hasAddressCity']['isPartOfDepartment']['insee']
                region = data['isLocatedAt'][0]['schema:address'][0]['hasAddressCity']['isPartOfDepartment']['isPartOfRegion']['rdfs:label']['fr']
                pays = data['isLocatedAt'][0]['schema:address'][0]['hasAddressCity']['isPartOfDepartment']['isPartOfRegion']['isPartOfCountry']['rdfs:label']['fr']
                latitude = data['isLocatedAt'][0]['schema:geo']['schema:latitude']
                longitude = data['isLocatedAt'][0]['schema:geo']['schema:longitude']
                adresse_complete = [adresse, cp, ville, departement, code_dpt, region, pays, latitude, longitude]
                liste_adresses.append(adresse_complete)
                
            except:
                adresse=""
                liste_adresses.append(adresse)
            

    df_adresses=pd.DataFrame(liste_adresses)
    
    df_adresses = df_adresses.rename({0:'Adresse', 1:'CP', 2:'Ville', 3:'Département', 4: 'N° Dpt', 5: 'Région', 6: 'Pays', 7:'Latitude', 8:'Longitude'}, axis='columns')
    
    return df_adresses
```

### df_ainelpé

```python

def NLP(dfindex, lien_avant_objects) :

    liste_NLP=[]

    for i in range(len(df1)) :
    #for i in range (4):

        with open(f"{lien_avant_objects}/objects/{df1['file'][i]}", "r") as fichier:
            contenu = fichier.read()

            # Chargement des données JSON
            data = json.loads(contenu)
            
            liste_inter = []
            # Description longue
            try:
                NLP = data['hasDescription'][0]['dc:description']['fr'][0]
                liste_inter.append(NLP)
                
            except:
                None
                
            # description courte
            try:
                NLP=data['hasDescription'][0]['shortDescription']['fr'][0]
                liste_inter.append(NLP)
                
            except:
                None
                
            # thème
            try:
                for j in range (len(data['hasTheme'])):
                    try:
                        NLP=data['hasTheme'][j]['rdfs:label']['fr'][0]
                        liste_inter.append("".join(NLP))
                    except:
                        None     
            except:
                None
                
                
            # équipements
            try:
                for k in range (len(data['hasFeature'])):
                    try:
                        NLP=data['hasFeature'][k]['features'][0]['rdfs:label']['fr']
                        liste_inter.append(" ".join(NLP))
                    except:
                        None
            except:
                None
            
            
            # nom
            try:
                NLP=data['rdfs:label']['fr'][0]
                liste_inter.append(NLP)
            except:
                None

            str_inter = " ".join(liste_inter)
            liste_NLP.append(str_inter)

    df_NLP=pd.DataFrame(liste_NLP)
    df_NLP = df_NLP.rename({0:'NLP'}, axis='columns')
    
    # LISTE TYPE (classifications)
    liste_type=[]

    for i in range(len(df1)) :
        with open(f"{lien_avant_objects}/objects/{df1['file'][i]}", "r") as fichier:
            contenu = fichier.read()

            # Chargement des données JSON
            data = json.loads(contenu)

            try:
                types= data['@type']
                liste_type.append(types)
                
            except:
                types=""
                liste_type.append(types)
    
    # CREATION DF AVEC LES DIFFERENTES CLASSIFICATIONS
    df_NLP['type']=liste_type
    df_NLP['type_str'] = df_NLP['type'].apply(lambda x: ' '.join(x))
    
    # GET DUMMIES
    df_dummies = df_NLP['type_str'].str.get_dummies(" ")
    
    # Suppression des colonnes contenant 'schema'
    df_filtered = df_dummies.drop([col for col in df_dummies.columns if 'schema' in col], axis=1)
    
    df_ainelpé = pd.merge(df_NLP,
             df_filtered,
             left_index=True,
             right_index=True)    
    
    return df_ainelpé
```

### df_paiements

```python
def paiements(dfindex, lien_avant_objects) :

    liste_paiements=[]

    for i in range(len(df1)) :
        with open(f"{lien_avant_objects}/objects/{df1['file'][i]}", "r") as fichier:
            contenu = fichier.read()
            
            # Chargement des données JSON
            data = json.loads(contenu)
            try:
                liste_pay=[]
                for j in range (len(data['offers'])):
                    liste_pmts=[]
                    for k in range (len(data['offers'][j]['schema:acceptedPaymentMethod'])):
                        paiements = data['offers'][j]['schema:acceptedPaymentMethod'][k]['rdfs:label']['fr'][0]
                        liste_pmts.append(paiements)
                    liste_pay.append(liste_pmts)
                liste_paiements.append(liste_pay[0])
                
            except:
                paiements=""
                liste_paiements.append(paiements)

    df_paiements=pd.DataFrame(liste_paiements)
    
    df_paiements = df_paiements.rename({0:'Modes de paiements'}, axis='columns')
    
    return df_paiements
    
```
